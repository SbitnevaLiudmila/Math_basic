```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-15, 15, 100)
y1 = np.cos(x)
y2 = np.cos(0.5 * x)
plt.plot(x, y1, label='k=1')
plt.plot(x, y2, label='k=0.5')

plt.title("Графики функции cos(kx)")
plt.legend(loc="lower right")

plt.show()
```
