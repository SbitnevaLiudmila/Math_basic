1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
import math
x = [10,10,10]
y = [0,0,-10]

def vector_length(num):
    return math.sqrt(num[0]**2 + num[1]**2 + num[2]**2)

print(vector_length(x))
print(vector_length(y))
```

3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
#окружность (x – x0)**2 + (y – y0)**2 = r**2 
#тогда y = y0 [+/-] SQRT(r**2 - (x – x0)**2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
#центр окружности
x0 = -1
y0 = 3
# радиус
r = 4

x = np.linspace(x0-r, x0 + r, 1000)
y1 = y0 + np.sqrt((r**2 - (x - x0)**2))
y2 = y0 - np.sqrt((r**2 - (x - x0)**2))

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
#эллипс x**2/a**2 + y**2/b**2 = 1
```

```python
a=4
b=3
x = np.linspace(-a, a, 1000)
y1 = b*np.sqrt(1 - (x**2)/(a**2))
y2 = -b*np.sqrt(1 - (x**2)/(a**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()
```

```python
# гипербола x**2/a**2 - y**2/b**2 = 1
```

```python
a=1
b=1
x = np.linspace(-10, 10, 5000)
y1 = b*np.sqrt((x**2/a**2)-1)
y2 = -b*np.sqrt((x**2/a**2)-1)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = 2*X+4*Y
Z2 = 2*X+4*Y - 50
ax.plot_wireframe(X, Y, Z, linewidth=2, color='y')
ax.plot_surface(X, Y, Z2, color='grey')
show()
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 5, 501)
plt.plot(x,(1-(1-np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def eq(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(eq,(-2, 1))
print (x1, y1)
x1, y1 = fsolve(eq,(2, 5))
print (x1, y1)
x1, y1 = fsolve(eq,(4, 15))
print (x1, y1)
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1 

```python
# ответом будет множество точек, лежащих на y = x2 – 1 и находящихся строго выше графика exp(x) + x∙(1 – y) > 1 
# это кусок слева до х = -1.581835352893692 и кусок справа между х = 2.6181455730830665 и 4.200105841149688
# выделила эти куски красным цветом
```

```python
x = np.linspace(-4, 5, 501)
plt.plot(x,(1-(1-np.exp(x))/x))
plt.plot(x, x**2 - 1)
x_ = np.linspace(-4, -1.581835352893692, 501)
plt.plot(x_, x_**2 - 1, color = 'red')
x__ = np.linspace(2.6181455730830665, 4.200105841149688, 501)
plt.plot(x__, x__**2 - 1, color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
